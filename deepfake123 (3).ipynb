```python
import os
import numpy as np
import torch
import torchvision.transforms as transforms
from transformers import Wav2Vec2Processor, Wav2Vec2Model, ViTFeatureExtractor, ViTModel
import torchaudio
from sklearn.metrics import silhouette_score, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import cv2
from PIL import Image
import json
import umap
import hdbscan
from hdbscan import approximate_predict
from collections import Counter
import joblib

dataset_path = "/kaggle/input/localized-audio-visual-deepfake-dataset-lav-df/LAV-DF"
train_path = os.path.join(dataset_path, "train")
metadata_path = os.path.join(dataset_path, "metadata.json")
output_path = "/kaggle/working"


with open(metadata_path, "r") as f:
    metadata = json.load(f)
print(f"✅ Metadata Loaded! Total Videos: {len(metadata)}")
```

```python
print("🚀 Loading Models...")
vit_path = "/kaggle/input/vit-base-patch16-224"
feature_extractor = ViTFeatureExtractor.from_pretrained(vit_path)
vit_model = ViTModel.from_pretrained(vit_path)
vit_model.eval()

processor = Wav2Vec2Processor.from_pretrained("/kaggle/input/wav2vec22/wav2vec2")
wav2vec2_model = Wav2Vec2Model.from_pretrained("/kaggle/input/wav2vec22/wav2vec2")
wav2vec2_model.eval()
print("Models Loaded Successfully!")
```

```python
vit_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])
```

```python
def extract_frames(video_path, frame_interval=30):
    frames = []
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if frame_count % frame_interval == 0:
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frames.append(Image.fromarray(frame))
        frame_count += 1
    
    cap.release()
    return frames
```

```python
def extract_visual_features_vit(video_path):
    frames = extract_frames(video_path)
    if not frames:
        print("❌ No Frames Extracted")
        return np.array([])

    with torch.no_grad():
        inputs = feature_extractor(images=frames, return_tensors="pt")
        features = vit_model(**inputs).last_hidden_state.mean(dim=1).squeeze().detach().numpy()

    return np.mean(features, axis=0) if features.size > 0 else np.array([])
```

```python
def extract_audio_features(video_path):
    audio_path = "/kaggle/working/audio.wav"
    command = f"ffmpeg -i {video_path} -q:a 0 -map a -ar 16000 -ac 1 {audio_path} -y"
    os.system(command)

    if not os.path.exists(audio_path):
        print("❌ Audio extraction failed.")
        return np.array([])

    waveform, sample_rate = torchaudio.load(audio_path)
    inputs = processor(waveform.squeeze().numpy(), sampling_rate=16000, return_tensors="pt", padding=True)

    with torch.no_grad():
        features = wav2vec2_model(**inputs).last_hidden_state.squeeze().numpy()

    return np.mean(features, axis=0) if features.size > 0 else np.array([])
```

```python
def get_actual_labels(video_ids):
    actual_labels = []
    for vid in video_ids:
        label = next((1 if data['n_fakes'] > 0 else 0 for data in metadata if data['file'].endswith(vid)), None)
        if label is not None:
            actual_labels.append(label)
        else:
            print(f"🚫 Warning: Video {vid} not found in metadata. Skipping...")
    return actual_labels
```

```python
def process_videos(video_list):
    multimodal_embeddings = []
    processed_videos = []

    for video_path in video_list:
        video_id = os.path.basename(video_path)
        print(f"🔍 Processing: {video_id}")

        video_features = extract_visual_features_vit(video_path)
        if video_features.shape[0] == 0:
            print("❌ No video features extracted. Skipping...")
            continue

        audio_features = extract_audio_features(video_path)
        if audio_features.shape[0] == 0:
            print("❌ No audio features extracted. Skipping...")
            continue

        multimodal_embedding = np.concatenate([video_features, audio_features])
        multimodal_embeddings.append(multimodal_embedding)
        processed_videos.append(video_id)

    return np.array(multimodal_embeddings), processed_videos
```

```python
video_paths = sorted([os.path.join(train_path, f) for f in os.listdir(train_path) if f.endswith(".mp4")])
train_videos, test_videos = video_paths[:700], video_paths[700:800]


train_embeddings, processed_train_ids = process_videos(train_videos)
test_embeddings, processed_test_ids = process_videos(test_videos)


train_labels = get_actual_labels(processed_train_ids)
test_labels = get_actual_labels(processed_test_ids)
```

```python
import numpy as np
from sklearn.metrics import silhouette_score, davies_bouldin_score
from collections import Counter

def calculate_purity(labels, cluster_labels):
    unique_clusters = np.unique(cluster_labels)
    total_samples = len(labels)
    total_correct = 0

    for cluster in unique_clusters:
        if cluster == -1:  
            continue
        cluster_indices = np.where(cluster_labels == cluster)[0]
        true_labels = labels[cluster_indices]
        most_common_label = Counter(true_labels).most_common(1)[0][1]
        total_correct += most_common_label

    purity = total_correct / total_samples
    return purity


train_valid_idx = cluster_labels_train != -1
train_silhouette = silhouette_score(reduced_train_features[train_valid_idx], cluster_labels_train[train_valid_idx])
train_davies_bouldin = davies_bouldin_score(reduced_train_features[train_valid_idx], cluster_labels_train[train_valid_idx])

test_valid_idx = cluster_labels_test != -1
test_silhouette = silhouette_score(reduced_test_features[test_valid_idx], cluster_labels_test[test_valid_idx])
test_davies_bouldin = davies_bouldin_score(reduced_test_features[test_valid_idx], cluster_labels_test[test_valid_idx])
import numpy as np

train_labels = np.array(train_labels)
test_labels = np.array(test_labels)


train_purity = calculate_purity(train_labels, cluster_labels_train)
test_purity = calculate_purity(test_labels, cluster_labels_test)

print(f"📊 Silhouette Score (Train): {train_silhouette:.4f}")
print(f"📊 Silhouette Score (Test): {test_silhouette:.4f}")
print(f"📊 Davies-Bouldin Index (Train): {train_davies_bouldin:.4f}")
print(f"📊 Davies-Bouldin Index (Test): {test_davies_bouldin:.4f}")
print(f"✅ Cluster Purity (Train): {train_purity * 100:.2f}%")
print(f"✅ Cluster Purity (Test): {test_purity * 100:.2f}%")

print(f"🔎 Cluster Distribution (Train): {Counter(cluster_labels_train)}")
print(f"🔎 Cluster Distribution (Test): {Counter(cluster_labels_test)}")

```

```python
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python

smote = SMOTE(random_state=42)

train_embeddings_resampled, train_labels_resampled = smote.fit_resample(train_embeddings, train_labels)

print("✅ After SMOTE - Class Distribution:")
print(f"Class 0: {sum(train_labels_resampled == 0)}")
print(f"Class 1: {sum(train_labels_resampled == 1)}")
```

```python
def evaluate_model(model, train_embeddings, train_labels, test_embeddings, test_labels):
    model.fit(train_embeddings, train_labels)
    predictions = model.predict(test_embeddings)

    print("🔎 Evaluation Metrics:")
    print(f"Accuracy: {accuracy_score(test_labels, predictions):.4f}")
    print(f"Precision: {precision_score(test_labels, predictions):.4f}")
    print(f"Recall: {recall_score(test_labels, predictions):.4f}")
    print(f"F1 Score: {f1_score(test_labels, predictions):.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(test_labels, predictions))
```

```python
print(" Training Logistic Regression with SMOTE...")
logistic_model = LogisticRegression(random_state=42)
evaluate_model(logistic_model, train_embeddings_resampled, train_labels_resampled, test_embeddings, test_labels)

print(" Training Support Vector Machine with SMOTE...")
svm_model = SVC(kernel='linear', random_state=42)
evaluate_model(svm_model, train_embeddings_resampled, train_labels_resampled, test_embeddings, test_labels)
```
